# Case-Study 1

```python
df = pd.read_csv('Datasets/customer_churn.csv')
```

### Sam has to build a bar-plot for the ‘Contract’ column:
* Set the x-axis label to be ‘Contract Type of customer’
* Set the y-axis label to be ‘Count’
* Set the title of the plot to be ‘Distribution of Contract’
* Assign ‘orange’ color to all the bars

```python
contract_counts = df['Contract'].value_counts()

plt.bar(contract_counts.index, contract_counts, color='orange', edgecolor='black')

plt.title("Distribution of Contract", fontsize=20)
plt.xlabel("Contract Type of Customer")
plt.ylabel("Count")
plt.show()
```

### Sam has to build a histogram for the ‘MonthlyCharges’ column:
* Set the x-axis label to be ‘Monthly Charges Incurred’
* Set the y-axis label to be ‘Count’
* Set the title of the plot to be ‘Distribution of Monthly Charges’
* Assign ‘forestgreen’ color to the bins

```python
plt.hist(df['MonthlyCharges'], bins=10, color='forestgreen', edgecolor='black')

plt.title("Distribution of Monthly Charges", fontsize=20)
plt.xlabel("Monthly Charges Incurred")
plt.ylabel("Count")
plt.show()
```

### Sam has to build a scatter-plot between ‘TotalCharges’ and ‘tenure’. ‘TotalCharges’ should be on the y-axis and ‘tenure’ should be on the x-axis.
* Set the x-axis label to be ‘Tenure of the customer’
* Set the y-axis label to be ‘Total chargesIncurred’
* Set the title of the plot to be ‘Total Charges vs Tenure’
* Assign ‘indigo’ color to the points

```python
plt.scatter(df['tenure'], df['TotalCharges'], color='indigo')

plt.title('Total Charges vs Tenure', fontsize=20)
plt.xlabel('Tenure of the customer')
plt.ylabel('Total charges Incurred')
plt.show()
```

### Sam has to build a box-plot between ‘MonthlyCharges’ and ‘PaymentMethod’. ‘MonthlyCharges’ should be on the y-axis and ‘PaymentMethod’ should be on the x-axis.
* Set the x-axis label to be ‘Payment Method of customer’
* Set the y-axis label to be ‘Monthly ChargesIncurred’
* Set the title of plot to be ‘Monthly Charges vs. Payment Method’
* Assign ‘olive’ color to the box-plots

```python
sns.boxplot(y='MonthlyCharges', x='PaymentMethod', data=df, color='olive')

plt.title('Monthly Charges vs Payment Method', fontsize=20)
plt.xlabel('Payment Method of customer')
plt.ylabel('Monthly Chargesincurred')
plt.xticks(rotation=45)
plt.show()
```

# Case-Study 2

```python
cardf = pd.read_csv('Datasets/cars.csv')
cardf.head(1)
```

### Create a bar plot between number of cylinders and frequency of cars with that many number of cylinders
* Set xlabel as Number of cylinders
* Set ylabel as Frequency of cars
* Draw a bar plot

```python
cylinder = cardf['cyl'].value_counts()

plt.bar(cylinder.index, cylinder)

plt.title('Frequency of Cars by Number of Cylinders', fontsize=18)
plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of cars')
plt.show()
```

### Write code to load data from cars and print a bar graph of count of columns with null values

```python
nullval = cardf.isnull().sum()

plt.bar(nullval.index, nullval)

plt.xticks(rotation=90)
plt.show()
```

### Use the 'mpg' (Miles Per Gallon column) and draw a histogram:
* Set xlabel: Miles per gallon
* Set ylabel: Frequency
* Set title as Miles Per Gallon Histogram
* Use mpg column to generate a histogram

```python
plt.hist(cardf['mpg'])

plt.title("Miles per Gallon Histogram", fontsize=18)
plt.xlabel("Miles per Gallon")
plt.ylabel("Frequency")
plt.show()
```

### Draw a boxplot on the cars dataframes hp column:
* Set xlabel: Car Horsepower
* Set title as Boxplot for car horsepower
* Use hp column to generate a boxplot

```python
plt.boxplot(cardf['hp'])

plt.title("Boxplot for Car HorsePower", fontsize=18)
plt.xlabel("Car HorsePower")
plt.show()
```
