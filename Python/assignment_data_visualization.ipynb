# Assignment 2

### Generate a bar plot graph for the columns ‘carbs’ and ‘gear’:
* Map the ‘carbs’ onto the x-axis
* Map the ‘gear’ onto the y-axis
* Provide the x-axis label as Number of carburetors
* Provide the y-axis label as Number of forward gears
* Set the title as carbs vs gear

```python
plt.bar(cars.carb, cars.gear)

plt.title('Carbs vs Gear', fontname='Rockwell', fontsize=20)
plt.xlabel('Number of Carburetors', fontname='Arial', fontsize=14)
plt.ylabel('Number of Forward Gears', fontname='Arial', fontsize=14)
plt.show()
```

# Assignment 3

### Plot a histogram for the column ‘wt’:
* Map the ‘wt’ onto the x-axis
* Provide the x-axis label as ‘weight of the cars’
* Provide the y-axis label as ‘Count’
* Set the number of bins as 30
* Set the title as ‘Histogram for the weight values’

```python
plt.hist(cars.wt, bins=30)

plt.title('Histogram for the weight values', fontname='Rockwell', fontsize=20)
plt.xlabel('Weight of the cars', fontname='Arial', fontsize=14)
plt.ylabel('Count', fontname='Arial', fontsize=14)
plt.show()
```
