Q12. Create a list of the first 50 even natural numbers, and perform the following operations.

        a. Count the number of elements in the list.
        b. Reverse the sequence of the list.
        c. Sort the list in ascending and descending order.
        d. Get the index value for the element 44, and update the element with the number 100.
        e. Return a copy of the list, with the resulting list containing the square of each element.

```python
list_even = [2 * i for i in range(1,51)]

print("Original list is:", list_even)
```

```python
# Count the number of elements in the list

print(f"Lenghth of List is: {len(list_even)}")
```

```python
# Reverse the sequence of the list

rev_list_even = list_even[::-1]

print("Reversed list is:", rev_list_even)
```

```python
# Sort the list in ascending and descending order

asc_sort_list_even = sorted(list_even)

print(f"Sort the list in ascending order is:", asc_sort_list_even)

des_sort_list_even = sorted(list_even, reverse=True)

print(f"Sort the list in descending order is:", des_sort_list_even)
```

```python
# Get the index value for the element 44, & update the element with the number 100.

print("Index of 44:", list_even.index(44))

list_even[list_even.index(44)] = 100

print("Updated List is: ", list_even)
```

```python
# Return a copy of the list, with the resulting list 
# containing the square of each element.

square_list_even = [i ** 2 for i in list_even]

print("Square List of All elements is: ", square_list_even)
```

Q13. Create a nested dictionary and perform the following operations.
        
        a. Return a list with the key value pairs from the dictionary.
        b. Return a list with just the keys from the dictionary.
        c. Remove the first and last key value from the dictionary.
        d. Update the last key value pair in the dictionary after removing in the 3rd step.

```python
company_dict = {
    "Engineering": {
        "Alice Smith": "Software Engineer",
        "Age": 30,
        "Location": "New York"
        },
    "Marketing": {
        "Carol Davis": "Marketing Manager",
        "Age": 29,
        "Location": "Chicago"
        },
    "Sales": {
        "Eva Brown": "Sales Representative",
        "Age": 32,
        "Location": "Houston"
        },
    "Ceo": {
        "Bhautik Gondaliya": "CEO & MD",
        "Age": 23,
        "Location": "India"
    }
    }
```

```python
# Return a list with the key value pairs from the dictionary.

list_company_dict = []

for key1, value1 in company_dict.items():
    for key, val in value1.items():
        list_company_dict.append([key1, [key, val]])

print(f"Company Dictionary in List Form: {list_company_dict}")
```

```python
# Return a list with just the keys from the dictionary.

list_keys_company_dict = [i for i in company_dict.keys()]

print(f"Company Dictionary's Key is: {list_keys_company_dict}")
```

```python
# Remove the first and last key value from the dictionary.

if len(list_keys_company_dict) >= 2:

    first_key = list_keys_company_dict[0]
    last_key = list_keys_company_dict[-1]

    del company_dict[first_key]
    del company_dict[last_key]

elif len(list_keys_company_dict) == 2:
    company_dict.clear()
    
else:
    print("Not 2 or more keys in the Dictionary.")

print("Updated Company Dictionary is: ", company_dict)
```

```python
# Update the last key value pair in the dictionary after removing in the 3rd step.

list_keys_company_dict = [i for i in company_dict.keys()]

last_key = list_keys_company_dict[-1]

company_dict[last_key] = {
        "Bhautik Gondaliya": "CEO & MD",
        "Age": 23,
        "Location": "India"
    }

print("Updated Company Dictionary is: ", company_dict)
```
