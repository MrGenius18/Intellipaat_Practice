# Auto Differntiation in Pytorch

* x = 3a^3 - b^2
* dx/da = 9a^2
* dx/db = -2b

```python
a = tensor([1,2,3], dtype=torch.float, requires_grad=True)
b = tensor([4,5,6], dtype=torch.float, requires_grad=True)
```

```python
x = 3*a**3 - b**2
# dx/da = [9, 36, 81]
# dx/db = [-8, -10, -12]
```

```python
dummy = torch.ones_like(x) # create tensor like [1.,1.,1.]
x.backward(gradient=dummy) # find diff to the backward propogation using greadient decent technique
```

```python
a.grad, b.grad
```

```python
x = tensor([5], dtype=torch.float, requires_grad=True)

Q = torch.cos(x) + 1/(1+torch.exp(x))
# dQ/dx = -sin(x) + 1/(1+e^x)
```

```python
dummmy = torch.ones_like(x)
```

```python
Q.backward(gradient=dummmy)
```
