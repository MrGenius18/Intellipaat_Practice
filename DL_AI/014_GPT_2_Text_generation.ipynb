### Download GPT-2 model from huggingface

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("gpt2")

model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
# prompt = "Give me a recipe for create an apple pie"
prompt = "how many odi centuries has kohli made?"
```

```python
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
```

```python
generated_tokens = model.generate(input_ids, max_length=1024, do_sample=True, temperature=0.9) # temperature is show creativity into 0.0-1.0
```

```python
tokenizer.batch_decode(generated_tokens)[0]
```
