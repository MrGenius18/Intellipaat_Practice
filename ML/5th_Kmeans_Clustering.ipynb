# Unsupervised Learning 
    ---> No Target Variable / No Labelled data

* Cluster Center ===> it's the arithmetic mean of each of the datapoints of the clusters.

* Each data point will be closure to its own cluster center than the other cluster center

>
>

### Good cluster will have
* within cluster distance should be smaller
* Inter cluster distance should be larger

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Creat a Data-sets

```python
from sklearn.datasets import make_blobs

x, y_true = make_blobs(n_samples=500, centers=7, cluster_std=0.8, random_state=0)
```

```python
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

### Train a Model

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=4, random_state=56)
```

```python
kmean.fit(x)
y_pred = kmean.predict(x)
```

```python
y_pred
```

### visulization of model

```python
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap='RdYlBu_r')
plt.show()
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap='RdYlBu_r', s=60)

# Centroids of each Cluster
centers = kmean.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.3)

plt.show()
```

### Decide Number of K

```python
kmean.inertia_   # wcss ==> within cluster some of store / distance
```

```python
WCSS = []

for k in range(1, 10):
    kmean = KMeans(n_clusters=k, random_state=56)
    kmean.fit(x)
    y_pred = kmean.predict(x)
    WCSS.append(kmean.inertia_)
```

```python
WCSS
```

```python
plt.plot(WCSS, marker='s')
plt.show()
```
