# Data Pre-Processing

### Lable Encoding 
* Target object value --> Numerical value (data['Churn'])

```python
data['Churn'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Churn'] = le.fit_transform(data['Churn'])
data['Churn'].value_counts()
```

# Train Test Split Our Data

### Predict the churn on the basis of monthly charges

```python
X = data[['MonthlyCharges']] # if not work so use 2d array
Y = data['Churn'] # Target Var

X_dt = data[['tenure']] # In the using Decision Tree
X_rf = data[['tenure', 'MonthlyCharges']] # In the using RAndom Forest
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.35, random_state=1)
```

# Logistic Regression Model

### Crete a Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logistic_reg = LogisticRegression()
Logistic_reg.fit(x_train,y_train)
```

### Predict the Model

```python
y_train_pradict = Logistic_reg.predict(x_train)
y_test_pradict = Logistic_reg.predict(x_test)
```

### Check the Accurancy

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_test_pradict)
```

```python
accuracy_score(y_train, y_train_pradict)
```

# Decision Tree Model

```python
xdt_train, xdt_test, ydt_train, ydt_test = train_test_split(X_dt, Y, train_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree = decision_tree.fit(xdt_train, ydt_train)
```

```python
ydt_train_pred = decision_tree.predict(xdt_train)
ydt_test_pred = decision_tree.predict(xdt_test)
```

```python
accuracy_score(ydt_train, ydt_train_pred)
```

```python
accuracy_score(ydt_test, ydt_test_pred)
```

# Random Forest Model

```python
xrf_train, xrf_test, yrf_train, yrf_test = train_test_split(X_rf, Y, train_size=0.7, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

Random_Forest = RandomForestClassifier()
Random_Forest = Random_Forest.fit(xrf_train, yrf_train)
```

```python
yrf_train_pred = Random_Forest.predict(xrf_train)
yrf_test_pred = Random_Forest.predict(xrf_test)
```

```python
accuracy_score(yrf_train, yrf_train_pred)
```

```python
accuracy_score(yrf_test, yrf_test_pred)
```
