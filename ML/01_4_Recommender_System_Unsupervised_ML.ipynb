# Build a **“Recommender System”** using the **Singular Value Decomposition (SVD)** algorithm.


```python
import pandas as pd
import numpy as np
```

>

## Load the **‘ratings’** and **‘movies’** datasets which is a part of ‘MovieLense’


```python
ratings = pd.read_csv("Datasets/ratings.csv")
ratings.head()
```

```python
movies = pd.read_csv("Datasets/movies.csv")
movies.head()
```

>

## Find the unique number of users and movies in the ‘ratings’ dataset


```python
n_users = ratings.userId.unique().shape[0]
n_movies = ratings.movieId.unique().shape[0]
print(f'Number of users = {n_users} & Number of movies = {n_movies}')
```

>

## Create a rating matrix for the ‘ratings’ dataset and store it in ‘Ratings’


```python
Ratings = ratings.pivot(index="userId", columns="movieId", values="rating").fillna(0)
Ratings
```

>

## Load the ‘ratings’ dataset as SVD’s Dataset object and compute 3-fold cross-validation using the SVD object
* **!pip install scikit-surprise**

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
```

```python
reader = Reader()
svd = SVD()

# Load ratings dataset with Dataset library
data = Dataset.load_from_df(ratings[["userId", "movieId", "rating"]], reader)

# Compute the RMSE of the SVD algorithm.
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=3, verbose=True)
```

>

## Find all the movies rated as 5 stars by user id ‘5’ and store it in ‘ratings_1’ data frame


```python
ratings_1 = ratings[(ratings["userId"] == 5) & (ratings["rating"] == 5)]
ratings_1 = ratings_1.set_index("movieId")
ratings_1 = ratings_1.join(movies)["title"]
ratings_1.head()
```

>

## Create a shallow copy of the ‘movies’ dataset and store the result in ‘user_5’


```python
user_5 = movies.copy()
user_5.head()
```

>

## Train a recommender system using the SVD object and predict the ratings for user id ‘5’


```python
#create a training set for svd
trainset = data.build_full_trainset()
svd.fit(trainset)
```

```python
#Predict the ratings for user5
user_5['Estimate_Score'] = user_5['movieId'].apply(lambda x: svd.predict(1, x).est)

user_5 = user_5.drop(['genres'], axis = 1).sort_values('Estimate_Score', ascending=False).set_index("movieId")
user_5.head()
```
