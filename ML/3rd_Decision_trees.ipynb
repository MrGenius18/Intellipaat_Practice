```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('daily_weather.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
df['high_humidity_3pm'].value_counts()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
df.columns
```

```python
dependent_var = 'high_humidity_3pm'
```

```python
independent_var = ['air_pressure_9am', 'air_temp_9am', 'avg_wind_direction_9am',
                   'avg_wind_speed_9am', 'max_wind_direction_9am', 'max_wind_speed_9am',
                   'rain_accumulation_9am', 'rain_duration_9am', 'relative_humidity_9am']
```

```python
X = df[independent_var]
y = df[dependent_var]
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
humidity_classifier = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes = 10, random_state = 10)
humidity_classifier.fit(X_train, y_train)
```

```python
y_pred_train = humidity_classifier.predict(X_train)
y_pred_test = humidity_classifier.predict(X_test)
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
accuracy_score(y_test, y_pred_test)
```
