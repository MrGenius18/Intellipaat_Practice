```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
```

# **K-Means Clustering**

```python
df = pd.read_csv("customers.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
plt.scatter(df["Age"], df["Spending Score (1-100)"])
plt.show()
```

```python
k_mean = KMeans(n_clusters=3, random_state=42)
k_mean.fit(df[['Age', 'Spending Score (1-100)']])
df['Cluster'] = k_mean.labels_
```

```python
plt.scatter(df['Age'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis', s=50)

# Adding labels and title
plt.title('Customer Segmentation: K-Means Clustering')
plt.xlabel('Age')
plt.ylabel('Spending Score')

# Show the plot
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
WCSS = []

for k in range(1, 11):
    kmean = KMeans(n_clusters=k, random_state=42)
    kmean.fit(df[['Age', 'Spending Score (1-100)']])
    y_pred = kmean.predict(df[['Age', 'Spending Score (1-100)']])
    WCSS.append(kmean.inertia_)
```

```python
plt.plot(WCSS, marker="s")
plt.show()
```

# **Agglomerative Clustering**

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
df = df.data
```

```python
agg_clust = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
agg_clust.fit(df)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(df, method='ward')

# Plotting the dendrogram
dendrogram(linked)
plt.title("Dendrogram of Agglomerative Clustering (Iris Dataset)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```
